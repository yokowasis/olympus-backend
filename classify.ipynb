```python
# get all files on classification
import pandas.errors
import os
import glob
import pandas as pd

path = r'./classification' # use your path
all_files = glob.glob(os.path.join(path, "*.csv"))

column_names = ["spanish", "english", "sdgs", "ex1", "ex2","ex3"]

df_from_each_file = []
for f in all_files:
    try:
        df = pd.read_csv(f, encoding='ISO-8859-1', sep="|", names=column_names, skiprows=1)
        df_from_each_file.append(df)
    except pandas.errors.ParserError:
        print(f"Error parsing file: {f}")

df = pd.concat(df_from_each_file, ignore_index=True)

# merge column sdgs, ex1, ex2, ex3 into one column
df['sdgs'] = df['sdgs'].fillna('')
df['ex1'] = df['ex1'].fillna('')
df['ex2'] = df['ex2'].fillna('')
df['ex3'] = df['ex3'].fillna('')

df['sdgs'] = df['sdgs'].astype(str) + df['ex1'].astype(str) + df['ex2'].astype(str) + df['ex3'].astype(str)

# replace space with empty string in sdgs column
df['sdgs'] = df['sdgs'].str.replace(' ', '')


# drop ex1, ex2, ex3
df = df.drop(columns=['ex1', 'ex2', 'ex3'])

df.to_csv("all_articles.csv", index=False, encoding='utf-8-sig')
```
