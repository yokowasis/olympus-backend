```python
from db import execQuery, getQuery
import pandas as pd

df = pd.read_csv('./dataset/merged_dataset.csv', encoding='ISO-8859-1', sep=",")
df.head()
```

```python
# empty table
execQuery ("TRUNCATE TABLE main2", "")
for i in range(len(df)):
    row = df.iloc[i]
    title = str(row['Title'])
    text = str(row['Text'])
    class_name = str(row['Class'])
    subclass = str(row['Subclass'])
    summary = str(row['Summary'])
    summary_vec = str(row['Summary_Vector'])
    title_vec = str(row['Title_Vector'])
    related = "1"
    sql = "INSERT INTO main2 (id, title, text, summary, class, subclass, related, title_vector, summary_vector) VALUES (%s, %s,%s,%s,%s,%s,%s,%s,%s)"
    val = (i, title, text, summary, class_name,
           subclass, related, title_vec, summary_vec)
    execQuery(sql, val)
```
