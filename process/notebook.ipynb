```python
import pandas as pd

df = pd.read_csv(
    './dataset_spain_sdgs_diperbaiki.csv',
    encoding='ISO-8859-1', sep=","
)

# get vector for 5 top rows
titles = []
df.head()
```

```python
# for i in range(0, dfNew.shape[0]):
vectorTitles = [];
for i in range(0, df.shape[0]):
    text = df['Text'][i]
    title = df['Title'][i]
    # print(i, title)
    titleVec = convertToVec(title)
    vectorTitles.append(titleVec)

# add column to df
dfnew = df
dfnew['TitleVec'] = vectorTitles

dfnew.head()
```

```python
dfnew.to_csv('dataset_spain_sdgs_diperbaiki_withtitlevec.csv', index=False)
dftitleonly = dfnew[['TitleVec']]
dftitleonly.to_csv("./dataset/title.csv", index=False)
dftitleonly.head()
```

```python
df[['Text']].head()
dfnew = df

summaries = []

for i in range(0, df.shape[0]):
    text = df['Text'][i]
    summary = summarize(text)
    summaries.append(summary)

dfnew['Summary'] = summaries    
dfnew.to_csv("./dataset/summary.csv")
dfnew.head()
```
